```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# 데이터
x = np.arange(-1.0, 1.01, 0.01) # x축의 값을 대입 : 배열 x에 -1~1범위의 값이 0.01의 간격으로 대입(첫번째 인자 포함, 마지막 인자 미포함)
y = 3 * x # 일차함수, 배열 x에 있는 각 요소의 값에 대해 3 X x를 계산하고 그 답으로 배열 y를 생성(아래 코드와 동일)
# y = []
# for i in range(len(x)):
#     y.append(3 * x[i])

# 그래프를 그린다
plt.plot(x, y)
plt.grid(color='0.8')
plt.show()
```
