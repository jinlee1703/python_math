```python
%matplotlib inline
import matplotlib.pyplot as plt

# 데이터
x = list(range(1, 11))        # x의 값(1~10)
y = []                        # 빈 리스트를 생성한다
for i in range(10):          # x가 1~10까지
    y.append(3 * x[i] - 24)  # y = 3x-24를 계산한 답을 리스트에 추가한다

# 그래프
plt.plot(x, y)
plt.grid(color='0.8')
plt.show()
```
