```python
# matplotlib는 아나콘다에 포함되어 있음
# %matplotlib inline : 파이썬이 아닌 주피터 노트북에 포함된 명령, Jupyter Notebook을 작성한 셀과 동일한 셀에 그래프가 그려짐
%matplotlib inline                 
import matplotlib.pyplot as plt

# 데이터
x = [1, 2, 3, 4, 5, 6, 7]
y = [64.3, 63.8, 63.6, 64.0, 63.5, 63.2, 63.1]

# 그래프를 그린다
# matplotlib 패키지에 포함된 모듈이며 그래프나 점을 그리는 함수가 정의돼 있습니다.
plt.plot(x, y, marker='o', color='red')         # 꺾은선을 그린다
plt.grid(color='0.8')                           # 격자를 표시한다
plt.show()                                      # 화면에 표시한다
```
